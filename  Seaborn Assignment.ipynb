## 1

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Five common types of plots we can create using Seaborn, along with their typical use cases:

1. **Scatter Plot**:
   - **Use**: Scatter plots are used to visualize the relationship between two continuous variables. They are useful for identifying patterns, trends, or clusters in data. We can also use different colors or sizes to represent additional information, such as a third variable.

2. **Line Plot**:
   - **Use**: Line plots are used to display data points connected by lines. They are typically used for time series data, showing how a variable changes over time. Line plots help visualize trends and variations in data.

3. **Histogram**:
   - **Use**: Histograms are used to visualize the distribution of a single continuous variable. They display the frequency or count of data points in various bins or intervals. Histograms are useful for understanding the central tendency, spread, and shape of a dataset.

4. **Bar Plot**:
   - **Use**: Bar plots are used to compare categories or discrete values across different groups. They are suitable for displaying counts or frequencies of categories. We can create both vertical and horizontal bar plots.

5. **Box Plot (Box-and-Whisker Plot)**:
   - **Use**: Box plots are used to visualize the distribution of a dataset and identify potential outliers. They display summary statistics such as the median, quartiles, and the range of the data. Box plots are valuable for comparing the distribution of a variable between different categories or groups.

## 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# facing error while loading sns.load_dataset("fmri")--so loaded local copy
fmri = pd.read_csv('fmri.csv')

# Create a line plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Use the "lineplot" function to create the line plot with hue by "event" and style by "region"
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add labels and a legend
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time")
plt.legend(title="Event and Region")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# facing error while loading sns.load_dataset("titanic")--so loaded local copyt
titanic =pd.read_csv('titanic.csv')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot a box plot for 'age' by 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Age Distribution by Passenger Class')

# Plot a box plot for 'fare' by 'pclass'
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Fare Distribution by Passenger Class')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# facing error while loading sns.load_dataset("diamond")--so loaded local copy
diamonds = pd.read_csv('diamonds.csv')

# Create a histogram with hue for 'cut' column
sns.histplot(data=diamonds, x="price", hue="cut", element="step", common_norm=False)

# Set labels and a title
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Diamond Prices by Cut")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = pd.read_csv('flights.csv')

# Pivot the data for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu")  # Use fmt=".0f" to format as integers

# Set labels and a title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap")

# Show the heatmap
plt.show()
```
